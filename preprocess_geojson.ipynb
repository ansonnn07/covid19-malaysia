```python
state_dict = {'trg': 'TERENGGANU', 'jhr': 'JOHOR', 'kdh': 'KEDAH', 
              'ktn': 'KELANTAN', 'mlk': 'MELAKA', 'nsn': 'NEGERI SEMBILAN', 
              'phg': 'PAHANG', 'png': 'PENANG', 'prk': 'PERAK', 
              'pls': 'PERLIS', 'sbh': 'SABAH', 'srw': 'SARAWAK',
              'sgr': 'SELANGOR'}
```

```python
coords_ser = df.loc[df.state.str.contains('trg'), 'coord']
full_coords = []
for lst in coords_ser:
    full_coords.extend(lst)
    # print(lst); break
full_coords = np.expand_dims(full_coords, axis=0).tolist()
# full_coords
```

```python
np.array(full_coords).shape
```

```python
for key, stateName in state_dict.items():
    coords_ser = df.loc[df.state.str.contains(key), 'coord']
    full_coords = []
    for lst in coords_ser:
        full_coords.extend(lst)
        # print(lst); break
    full_coords = np.expand_dims(full_coords, axis=0).tolist()
    df = df.append({'type': 'Feature', 
                    'properties': {'name': stateName}, 
                    'geometry': {'type': 'Polygon', 
                                 'coordinates': full_coords},
                    'state': stateName
                   }, ignore_index=True)
```

```python
df = df.drop(df.loc[3:44].index).reset_index(drop=True)
```

```python
df
```

```python
def replace_type(geo_dict):
    geo_dict['type'] = 'Polygon'
    geo_dict['coordinates'] = np.expand_dims(geo_dict['coordinates'], axis=0).tolist()
    return geo_dict
```

```python
df.loc[:2, 'geometry'] = df.loc[:2, 'geometry'].apply(replace_type)
```

```python
df.drop(columns=['state', 'geotype', 'coord'], inplace=True)
```

```python
df
```

```python
# To save
filename = 'new_geojson.geojson
result = df.to_json(orient='records')
parsed = json.loads(result)
parsed_geojson = {'type': 'FeatureCollection', 'features': parsed}
with open(filename', 'w') as f:
    f.write(json.dumps(parsed_geojson, indent=2))
```

```python
new_df = pd.DataFrame(columns=['type', 'properties', 'geometry'])

debug = 0
for item in correct_df.itertuples():
    # full coordinate list with the right order
    full_coords = []
    for i in item.coord_order:
        curr_list = []
        for j in i:
            for idx, state_key in enumerate(j.split(',')):
                # get the row from the geojson df
                row = df.loc[df.state == state_key].copy()
            
                if len(row) == 0:
                    # state key not found, reverse it to find it
                    reversed_key = state_key[3:] + state_key[:3]
                    row = df.loc[df.state == reversed_key].copy()
                    coords = row.geometry.iloc[0]['coordinates']
                    # then reverse the coordinate's order too
                    # CAREFUL OF REVERSING THE ORIGINAL DF!
                    coords = coords.copy()
                    coords.reverse()
                    
                    if debug and item.Name == 'SELANGOR':
                        print("reversed_key: ", reversed_key)
                else:
                    coords = row.geometry.iloc[0]['coordinates']
                    
                    if debug and item.Name == 'SELANGOR':
                        print("state_key: ", state_key)
                
                # add to the current list of polygons for
                #  the current "state key", e.g. "pngocn"
                curr_list.extend(coords)
                
                if debug and item.Name == 'SELANGOR':
                    print("coords: ", coords)
                    
            if debug and item.Name == 'SELANGOR':
                print()
                print("curr_list: ", curr_list)
            
            # add to the current list of lists of polygons
            #  for the current state
            full_coords.append(curr_list)
    
    # add one more dimension for multipolygon
    full_coords = np.expand_dims(full_coords, axis=0).tolist()
    
    new_df = new_df.append({'type': 'Feature',
                            'properties': {'id': int(item.id), 'name': item.Name},
                            'geometry': {'type': 'MultiPolygon',
                                         'coordinates': full_coords}
                           }, ignore_index=True)
    # break
new_df
```
